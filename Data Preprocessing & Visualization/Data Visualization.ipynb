**Methodology for exploring the relationship between the fields**

*Generally used to explore relationship between response variable and explanatory variable in supervised learning*

---

|Explanatory variable | Response |  Plot |
|--------             |------    |----   |
|Continuous |Continuous |Scatter plot|
|Continuous |Categorical|Box plot, Density plot|

**When response and explanatory variable both continuous**

```python
#Import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the dataset
data = pd.read_excel('Chemical_Yield.xlsx')
```

```python
data.head()
```

```python
# Descriptive statistics
data.describe()
```

```python
# Check if there is any missing value
data.count()
```

*Missing value is present in the dataset- in Temperature feature*

```python
# We impute missing value of Temperature feature with mean
Temperature = data.Temperature
```

```python
# Compute the mean of Temperature
temp_mean = Temperature.mean()
```

```python
# Replace with mean
Temperature.fillna(temp_mean,inplace=True)
```

```python
# we clean up the data by removing SL.No field
data = data.iloc[:,1:5]
```

```python
#Data Visualization
sns.pairplot(data)
plt.show()
```
